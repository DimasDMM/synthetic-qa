# Group evaluations

Once we have produced several `xx_metrics.json` files, this notebook will put all together in a single CSV file.

```python
import glob
import json
import os
import pandas as pd
import re
```

```python
def group_metrics(ckpt_name):
    df = {
        'dataset_names': [],
        'exact_scores': [],
        'f1_scores': [],
    }
    
    input_path = '../artifacts/%s/predictions/' % ckpt_name
    output_file = '../artifacts/%s/predictions/all_metrics.csv' % ckpt_name

    files = glob.glob('%s*_metrics.json' % input_path)
    for file in files:
        filename = os.path.basename(file)
        dataset_name = re.sub(r'_metrics.json', '', filename)
        with open(file, 'r') as fp:
            scores = json.load(fp)

        df['dataset_names'].append(dataset_name)
        df['exact_scores'].append(scores['exact'])
        df['f1_scores'].append(scores['f1'])

    df = pd.DataFrame.from_dict(df).set_index('dataset_names')
    df.to_csv(output_file)

    return df
```

```python
group_metrics('qa_squad2')
```

```python
group_metrics('qa_synthetic')
```
